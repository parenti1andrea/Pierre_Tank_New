```python
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

TANK_RADIUS = 1 # cm 
```

```python
random_number = np.random.uniform(size=100000)

yy = TANK_RADIUS * (2*random_number - 1)
xx = np.sqrt(TANK_RADIUS*TANK_RADIUS - yy*yy)

plt.hist(yy,bins=np.linspace(-1,1,100))
plt.hist(xx,bins=np.linspace(-1,1,100),alpha=0.6)

plt.yscale('log')
```

```python
plt.scatter(xx,yy)
```

```python
cz = np.random.uniform(size=100000)
sz = np.sqrt(1 - cz*cz)
azi = np.random.uniform(0, 2*np.pi, size=100000)

cxy = - np.sqrt(1 - cz*cz)

x = (- sz * np.cos(azi) * xx + sz * np.sin(azi) * yy) / cxy
y = (- sz * np.cos(azi) * yy + sz * np.sin(azi) * xx) / cxy

plt.scatter(x,y)
plt.gca().set_aspect('equal')

```
